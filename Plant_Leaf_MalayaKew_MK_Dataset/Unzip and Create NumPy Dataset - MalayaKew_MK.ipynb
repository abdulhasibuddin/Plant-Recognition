###Assign Directories:

###Unzip Files:

```
unzip = False # True, False
```

```
if unzip == True:
    print("Unzipping...")
    with zipfile.ZipFile(path_to_zip_file+filename, 'r') as zip_ref:
        zip_ref.extractall(directory_to_extract_to)
        unzip = False
    print("Unzipping complete!")
else:
    print(f"Unzip = {unzip}")
```

Details (Source: http://web.fsktm.um.edu.my/~cschan/downloads_MKLeaf_dataset.html)

MalayaKew (MK) Leaf dataset was collected at the Royal Botanic Gardens, Kew, England. It consists of scan-like images of leaves from 44 species classes. This dataset is very challenging as leaves from different species classes have very similar appearance.

There are two(2) folders associated with the dataset and a ReadMe file:

    1. D1 dataset
        It consists of segmented leaf images with size 256 * 256 pixels.​​
        Number of training and testing images is 2288 and 528 respectively.
    2. Dataset Groundtruth
        It consists of cropped image patches of leaf with size 256 * 256 pixels.
        Number of training and testing images is 34672 and 8800 respectively.

###Prepare Train Datadet:

```
#dir_name = f"{directory_to_extract_to}MK/D2/train_patch/"
dir_name = f"{directory_to_extract_to}D2_Gray_crop_{width}x{height}p/train_patch/"
print(f"dir_name: {dir_name}")
classes = os.listdir(dir_name)
print(f"len(classes) = {len(classes)}\n")

list_train_imageset = []
list_train_label = []
count_total = 0
for aClass in tqdm(classes):
#for aClass in classes:
    class_dir = f"{dir_name}{aClass}"
    list_files = os.listdir(class_dir)
    class_label = aClass[5:]

    #print(f"\nCurrent class = {aClass}:")
    #for filename in tqdm(list_files):
    for filename in list_files:
        img_dir = f"{class_dir}/{filename}"
        #print(f"img_dir = {img_dir}")
        #print(f"os.listdir(img_dir) = {os.listdir(img_dir)}")
        img = cv2.imread(img_dir, 0)
        list_train_imageset.append(img)
        list_train_label.append(aClass[5:])
        #break

    count_total += len(list_files)
    #print(f"# {class_dir} = {num_files}")
    #break

print(f"Total in {dir_name} = {count_total}\n")

print(f"len(list_train_imageset) = {len(list_train_imageset)}")
print(f"len(list_train_label) = {len(list_train_label)}\n")
print(f"list_train_label = {list_train_label}\n")

print(f"list_train_imageset[0].shape = {list_train_imageset[0].shape}")
print(f"list_train_label = {list_train_label}\n")
```

```
np_train_imageset = np.array(list_train_imageset)
np_train_label = np.array(list_train_label)

print(f'np_train_imageset.shape = {np_train_imageset.shape}')
print(f'np_train_label.shape = {np_train_label.shape}\n')
```

```
save_filename_X = f"D2_Gray_crop_{width}x{height}_train_X.pkl"
save_filename_Y = f"D2_Gray_crop_{width}x{height}_train_y.pkl"

np.save(f"{directory_to_extract_to}{save_filename_X}", np_train_imageset, allow_pickle=True)
print(f"Saved: {directory_to_extract_to}{save_filename_X}")
np.save(f"{directory_to_extract_to}{save_filename_Y}", np_train_label, allow_pickle=True)
print(f"Saved: {directory_to_extract_to}{save_filename_Y}")
```

###Prepare Test Datadet:

```
dir_name = f"{directory_to_extract_to}D2_Gray_crop_{width}x{height}p/test_patch/"
print(f"dir_name: {dir_name}")
classes = os.listdir(dir_name)
print(f"len(classes) = {len(classes)}\n")

list_test_imageset = []
list_test_label = []
count_total = 0
for aClass in tqdm(classes):
    class_dir = f"{dir_name}{aClass}"
    list_files = os.listdir(class_dir)
    class_label = aClass[5:]

    for filename in list_files:
        img_dir = f"{class_dir}/{filename}"

        img = cv2.imread(img_dir, 0)
        list_test_imageset.append(img)
        list_test_label.append(aClass[5:])

    count_total += len(list_files)

print(f"Total in {dir_name} = {count_total}\n")

print(f"len(list_test_imageset) = {len(list_test_imageset)}")
print(f"len(list_test_label) = {len(list_test_label)}\n")
print(f"list_test_label = {list_test_label}\n")

print(f"list_train_imageset[0].shape = {list_train_imageset[0].shape}")
print(f"list_train_label = {list_train_label}\n")
```

```
np_test_imageset = np.array(list_test_imageset)
np_test_label = np.array(list_test_label)

print(f'np_test_imageset.shape = {np_test_imageset.shape}')
print(f'np_test_label.shape = {np_test_label.shape}\n')
```

```
save_filename_X = f"D2_Gray_crop_{width}x{height}_test_X.pkl"
save_filename_Y = f"D2_Gray_crop_{width}x{height}_test_y.pkl"

np.save(f"{directory_to_extract_to}{save_filename_X}", np_test_imageset, allow_pickle=True)
print(f"Saved: {directory_to_extract_to}{save_filename_X}")
np.save(f"{directory_to_extract_to}{save_filename_Y}", np_test_label, allow_pickle=True)
print(f"Saved: {directory_to_extract_to}{save_filename_Y}")
```
